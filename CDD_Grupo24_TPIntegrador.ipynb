##4.1.A Preparacion de los datos

####Correccion, limpieza y formateo de los datos

En el dataset no se encuentran datos faltantes, ni tampoco nulos

```python
df_unido.head(10)
```

```python
columnas_upper = [
    "salido_de",
    "ingresado_a",
    "provincia_origen",
    "departamento_origen",
    "provincia_destino",
    "departamento_destino"
]

for col in columnas_upper:
    if col in df_unido.columns:
        df_unido[col] = df_unido[col].astype(str).str.upper()
    else:
        print(f"No existe la columna: {col}")

df_unido
```

Normalizamos todas las celdas de texto a mayusculas para mejorar la manipulacion de los datos

```python
#A continuacion realizaremos una limpieza en las celdas de texto, quitamos acentos, espacios en blanco al comienzo o final y posibilidad de dobles espacios
columnas_ascii = [
    "salido_de",
    "ingresado_a",
    "provincia_origen",
    "departamento_origen",
    "provincia_destino",
    "departamento_destino"
]
for col in columnas_ascii:
    if col in df_unido.columns:
      df_unido[col] = (
          df_unido[col]
          .astype(str)
          .str.normalize('NFKD')
          .str.encode('ascii', errors='ignore')
          .str.decode('utf-8')
          .str.strip()
          .str.replace(r'\s+', ' ', regex=True))
    else:
        print(f"No existe la columna: {col}")

df_unido
```

Terminamos de limpiar las celdas de texto quitando posibles acentos, espacios en blanco al comienzo o al final, y tambien eliminamos los posibles doble espacios intermedios que puedan existir

```python
# buscamos fechas en formato extraño y las convertimos
# a un formato más entendible para el cliente

df_unido['fecha'] = pd.to_datetime(
    df_unido['fecha'],
    errors='coerce'
)

# mostramos fechas inválidas
fechas_invalidas = df_unido[df_unido['fecha'].isna()]
if not fechas_invalidas.empty:
    print("Fechas inválidas encontradas:")
    print(fechas_invalidas)
else:
    print("No se encontraron fechas inválidas.")


df_unido
```

Verificamos que el dataset no posee formatos de fechas invalidos

```python
def obtener_estacion(fecha):
  mes = fecha.month
  if mes in [12,1,2]:
    return 'VERANO'
  elif mes in [3,4,5]:
    return 'OTONIO'
  elif mes in [6,7,8]:
    return 'INVIERNO'
  elif mes in [9,10,11]:
    return 'PRIMAVERA'

df_unido['estacion'] = df_unido['fecha'].apply(obtener_estacion)
df_unido
```

####**Construir tablas de frecuencia y proporciones para las variables categóricas.**

```python
tabla_anio = pd.DataFrame({
    'frecuencia': (
        df_limpio['anio']
        .value_counts()
    ),

    'proporcion': (
        df_limpio['anio']
        .value_counts(normalize=True)
    ),
    'porcentaje': (
        df_limpio['anio']
        .value_counts(normalize=True) * 100
    ).round(2)
})

tabla_anio
```

Creamos una tabla de frecuencia por anios.
Podemos ver que los movimientos comprendidos entre lso anio 2015 a 2018 han sido parejos, mostrando un leve aumento del 1% para anios 2017 y 2018 con respecto de sus anteriores.

```python
tabla_estacion = pd.DataFrame({
    'frecuencia': (
        df_limpio['estacion']
        .value_counts()
    ),

    'proporcion': (
        df_limpio['estacion']
        .value_counts(normalize=True)
    ),

    'porcentaje': (
        df_limpio['estacion']
        .value_counts(normalize=True) * 100
    ).round(2)
})

tabla_estacion
```

Creamos una tabla de frecuencia por estaciones.
Podemos ver que el mayor porcentaje de movimientos se realiza durante el mes de Invierno, pero que sigue teniendo una distribucion pareja en porcentajes con el resto de las estaciones a lo largo de los anios.


```python
tabla_estacion_por_anio = pd.DataFrame({
    'frecuencia': (
        df_limpio.groupby(['anio', 'estacion']).size()
    ),

    'proporcion': (
        df_limpio.groupby(['anio', 'estacion']).size() / df_limpio.groupby('anio').size()
    ),
    'porcentaje': (
        df_limpio.groupby(['anio', 'estacion']).size() / df_limpio.groupby('anio').size() * 100
    ).round(2)
})

tabla_estacion_por_anio
```
